```python
# Initialized imports
import requests
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import altair as alt
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from pathlib import Path
```

### Fetching Cryptocurrency Data

```python
# Used the following endpoint to fetch json data
url = "https://min-api.cryptocompare.com/data/all/coinlist"
coinlist_data = requests.get(url).json()
```

```python
# Created a DataFrame 
coinlist_df = pd.DataFrame(coinlist_data["Data"]).T
coinlist_df.head()
```

```python
# Initialized provided csv file.
file_path = Path("Resources/crypto_data.csv")

# Created a DataFrame for the provided csv.
crypto_df = pd.read_csv(file_path)
crypto_df
```

### Data Preprocessing

```python
# Kept only necessary columns:
# 'CoinName','Algorithm','IsTrading','ProofType','TotalCoinsMined','TotalCoinSupply'
crypto_df.set_index("Unnamed: 0", inplace = True)
crypto_df.index.name = None
crypto_df.head()
```

```python
# Kept only cryptocurrencies that are trading
crypto_df = crypto_df[crypto_df["IsTrading"] == True]
crypto_df.head()
```

```python
# Kept only cryptocurrencies with a working algorithm
crypto_df = crypto_df[crypto_df['Algorithm'].isnull() == False]
crypto_df.head()
```

```python
# Removed the 'IsTrading' column
crypto_df.drop(columns=['IsTrading'], inplace=True)
crypto_df.head()
```

```python
# Removed rows with at least 1 null value
crypto_df.dropna(inplace=True)
```

```python
# Removed rows with cryptocurrencies having no coins mined
crypto_df = crypto_df[crypto_df["TotalCoinsMined"] > 0]
crypto_df.head()
```

```python
# Dropped rows where there are 'N/A' text values
crypto_df.drop(crypto_df.index[(crypto_df['CoinName'] == 'N/A') |
           (crypto_df['Algorithm'] == 'N/A') |
           (crypto_df['ProofType'] == 'N/A') |
           (crypto_df['TotalCoinsMined'] == 'N/A') | 
           (crypto_df['TotalCoinSupply'] == 'N/A')], inplace = True)
crypto_df.size
```

```python
# Stored the 'CoinName'column in its own DataFrame prior to dropping it from crypto_df 
coin_names_df = crypto_df[["CoinName"]]
coin_names_df.head()
```

```python
# Dropped the 'CoinName' column since it's not going to be used on the clustering algorithm
crypto_df.drop("CoinName", axis=1, inplace=True)
crypto_df.head()
```

```python
# Created dummy variables for text features
LE = LabelEncoder()
X = crypto_df.copy()
X['Algorithm'] = LE.fit_transform(X['Algorithm']) 
X['ProofType'] = LE.fit_transform(X['ProofType']) 
X.head()
```

```python
# Standardized data
X_scaled = StandardScaler().fit_transform(X)
print(X_scaled[0:5])
```

### Reducing Dimensions Using PCA

```python
# Used PCA to reduce dimensions to 3 principal components
pca = PCA(n_components=3)
```

```python
# Created a DataFrame with the principal components data
X_pca = pca.fit_transform(X_scaled)
pcs_df = pd.DataFrame(data= X_pca, index=X.index, columns = ["PC 1", "PC 2","PC 3"])
pcs_df.head(10) 

# Examined the explained variance
pca.explained_variance_ratio_
```

### Clustering Crytocurrencies Using K-Means

#### Finding the Best Value for `k` Using the Elbow Curve

```python
inertia = []
k = list(range(1, 11))

# Calculated the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)


# Created the Elbow Curve using hvPlot

elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
alt.Chart(df_elbow).mark_line().encode(x='k', y='inertia').properties(
    title='The elbow method for determining best K value',
    width=400
)
```

#### I selected the value of k at the “elbow” ie the point after which the distortion/inertia start decreasing in a linear fashion

Running K-Means with k=6

```python
# Initialized the K-Means model
k=6
model = KMeans(n_clusters=k, random_state=0)

# Fitted the model
model.fit(pcs_df)

# Predicted clusters
predictions = model.predict(pcs_df)

# Created a new DataFrame including predicted clusters and cryptocurrencies features

pcs_df["Class"] = model.labels_

pcs_df.head(10)
```

```python
# Created a new DataFrame named clustered_df, that includes the following columns "Algorithm", "ProofType", "TotalCoinsMined", "TotalCoinSupply", "PC 1", "PC 2", "PC 3", "CoinName", "Class" for visualization
clustered_df = crypto_df.merge(pcs_df, how='outer', left_index=True, right_index=True)
clustered_df['CoinName'] = coin_names_df.values
reorder = ['Algorithm', 'ProofType', 'TotalCoinsMined', 'TotalCoinSupply', 'PC 1', 'PC 2', 'PC 3', 'CoinName', 'Class']
clustered_df = clustered_df.reindex(columns=reorder)
clustered_df.head(10)
```

### Visualizing Results

#### 3D-Scatter with Clusters

```python
# Created a 2D-Scatter with the PCA data and the clusters
fig = alt.Chart(clustered_df).mark_circle(
    size=100
).encode(
    alt.X("PC 1", title="PC 1", scale=alt.Scale(zero=False)),
    alt.Y("PC 2", title="PC 2", scale=alt.Scale(zero=False)),
    color="Class",
    tooltip=["CoinName", "Algorithm", "TotalCoinsMined", "TotalCoinSupply"]
).properties(
    height=300,
    width=500,
    title="PC 1 vs PC 2 Color-encoded by Class").interactive()
fig
```
